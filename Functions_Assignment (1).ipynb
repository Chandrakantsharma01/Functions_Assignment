1. What is the difference between a function and a method in Python?

Answer: In Python, both functions and methods are blocks of reusable code designed to perform specific tasks, but their key distinction lies in their relationship with classes and objects:

**Function**:

- A function is a standalone block of code that is defined independently, not within a class.
- It can be called directly by its name from anywhere in the program where it is in scope.
- Functions typically operate on data passed to them as arguments and do not inherently have access to or modify the state of any specific object unless that object is explicitly passed as an argument.

**Method**:

- A method is a function that is defined within a class.
-It is associated with an object (an instance of that class) or the class itself.
-Methods are called on an object using dot notation (e.g., object_name.- - method_name()) or on the class itself (e.g., ClassName.method_name()).
- Methods defined for an object implicitly receive the object itself as their first argument (conventionally named self for instance methods). This allows methods to access and modify the object's attributes and other methods, encapsulating behavior related to that object.
- Methods can be categorized as instance methods, class methods, or static methods, each with slightly different ways of interacting with the object or class.

2. Explain the concept of function arguments and parameters in Python.

Answer: In Python, the terms "parameters" and "arguments" are distinct concepts related to functions.

Parameters are variables defined in the function's signature during its definition. They act as placeholders for the values that the function expects to receive when it is called. Parameters specify the types and number of inputs a function can accept.

Consider the following function definition:

    def greet(name, age):
    print(f"Hello, {name}! You are {age} years old.")
In this example, name and age are the parameters of the greet function.

Arguments are the actual values that are passed to a function when it is called. These values are assigned to the corresponding parameters within the function's scope, allowing the function to perform its operations using specific data.

    greet("Alice", 30)
Here, "Alice" and 30 are the arguments passed to the greet function. "Alice" is assigned to the name parameter, and 30 is assigned to the age parameter.

4. What is the purpose of the `return` statement in a Python function?

Answer: The return statement in a Python function serves two primary purposes:

**Exiting the Function**:

When a return statement is encountered during the execution of a function, it immediately terminates the function's execution. No further lines of code within that function are executed after the return statement. Control is then transferred back to the point in the program where the function was called.

**Returning a Value**:
The return statement can be used to send one or more values back to the caller of the function. This allows the function to produce an output that can be used by other parts of the program. The value(s) returned can be assigned to a variable, used in an expression, or passed as an argument to another function. If no value is specified with the return statement, or if the function reaches its end without encountering a return statement, the function implicitly returns None.

**Example**:

    def add_numbers(a, b):
        result = a + b
        return result  # Returns the sum of a and b

    sum_value = add_numbers(5, 3)
    print(sum_value) # Output: 8




5. What are iterators in Python and how do they differ from iterables?

Answer: In Python, an iterable is an object that can be looped over, such as a list, tuple, or string. An iterator, on the other hand, is an object that produces the next value in a sequence when you call its __next__() method. The key difference is that iterables are objects you can iterate through, while iterators are objects that manage the iteration process.

Difference between iterables & iterators in python ...

**Iterables**:

- An iterable is any object that can be used in a for loop or other iteration contexts.
- Examples include lists, tuples, strings, dictionaries, and sets.
-They provide a way to access their elements one at a time.
-They don't necessarily keep track of their position during iteration.
-To get an iterator from an iterable, you use the iter() function.

**Iterators**:

- An iterator is an object that remembers its position during iteration and knows how to get the next value.
- They implement the __next__() method, which returns the next item in the sequence.
- When there are no more items to return, __next__() raises a StopIteration exception.
- Iterators are also iterables, meaning they have an __iter__() method that returns themselves.
- Examples include the iterator returned by iter(some_iterable).

**In essence**:

- An iterable is a container that holds items.
- An iterator is a tool that steps through the items of a container.
- You can get an iterator from an iterable, but not all iterables are iterators.
- Every iterator is also an iterable, but not vice versa.


6. Explain the concept of generators in Python and how they are defined.

Answer: In Python, a generator is a special type of iterator that produces a sequence of values on demand, rather than storing them all in memory at once. This makes them memory-efficient, especially when dealing with large datasets or infinite sequences. Generators are defined using functions and the yield keyword.

1. Defining a Generator Function:
- A generator function is defined like a regular function, but it uses the yield keyword instead of return.
- When a generator function is called, it doesn't execute the code immediately. - Instead, it returns a generator object.
- The generator object is an iterator that can be used to produce the sequence of values defined in the function.
- Each time the next() function is called on the generator object, the function executes until it encounters a yield statement, which returns the yielded value and pauses the function's execution.
- The generator function's state is preserved between calls to next(), allowing it to continue from where it left off.

**Example**:


    def my_generator():
        yield 1
        yield 2
        yield 3

    gen = my_generator()
    print(next(gen))  # Output: 1
    print(next(gen))  # Output: 2
    print(next(gen))  # Output: 3
    #print(next(gen)) # Raises StopIteration error, as there are no more values to yield
2. Generator Expressions:

- Python also provides a concise way to create generators using generator expressions, which are similar to list comprehensions but use parentheses instead of square brackets.
- Example:(x*2 for x in range(10)) creates a generator that yields the double of each number from 0 to 9.
- Generator expressions are evaluated lazily, meaning the calculations are only performed when the values are requested.

**Key Benefits of Generators**:
*Memory Efficiency*: Generators don't store the entire sequence in memory at once, making them ideal for large or infinite sequences.
*Improved Performance*: Generators can improve performance by calculating values only when needed.
*Simplified Iteration*: Generators simplify the process of creating iterators, making it easier to work with sequences of data.


7. What are the advantages of using generators over regular functions?

Answer: Generators offer significant advantages over regular functions, primarily in memory efficiency and the ability to handle potentially infinite sequences. They achieve this by yielding values one at a time, on demand, rather than computing and storing an entire sequence in memory at once.

**Memory Efficiency**:

Regular functions typically return a complete data structure (like a list) holding all the results. Generators, on the other hand, use yield to produce values one at a time, significantly reducing memory consumption, especially when dealing with large or infinite sequences.

**Lazy Evaluation**:

Generators employ lazy evaluation, meaning they compute and return the next value only when it's requested. This is in contrast to regular functions, which compute everything upfront.

**Flexibility and Infinite Sequences**:

Generators can represent infinite sequences or very large datasets without running into memory limitations, as they only compute values as needed.

**Improved Readability**:

Generators can simplify code for iterative logic, making it more concise and easier to understand, especially when dealing with complex iteration patterns.

**Reusability**:
   
Gen erators can be easily composed and reused across different parts of an application, promoting modularity.

**Streamlined Asynchronous Code**:

Generators can be combined with async/await in languages like JavaScript to create cleaner and more readable asynchronous workflows.

9. Explain the purpose and usage of the `map()` function in Python.

Answer: The map() function in Python is a built-in function used to apply a specified function to each item in an iterable (like a list, tuple, or string) and returns a map object, which is an iterator containing the results.

**Purpose**:

The primary purpose of map() is to perform element-wise transformations on iterables efficiently, often replacing explicit for loops for cleaner and more concise code. It's particularly useful when you need to apply the same operation to every item in a collection.

**Usage**:

The syntax for map() is:

    map(function, iterable, ...)

**function**:

This is the function that will be applied to each item of the iterable(s). It can be a regular named function, a lambda function, or even a built-in function.

**iterable**:

This is one or more iterable objects whose elements will be passed as arguments to the function. If multiple iterables are provided, the function must accept a corresponding number of arguments, and map() will process elements in parallel until the shortest iterable is exhausted.

**Example**:

    # Using a named function
    def square(number):
        return number * number

    numbers = [1, 2, 3, 4]
    squared_numbers_map = map(square, numbers)
    # Convert the map object to a list to view the results
    print(list(squared_numbers_map))

    # Using a lambda function
    doubled_numbers_map = map(lambda x: x * 2, numbers)
    print(list(doubled_numbers_map))

    # Using multiple iterables
    list1 = [1, 2, 3]
    list2 = [4, 5, 6]
    sum_lists_map = map(lambda x, y: x + y, list1, list2)
    print(list(sum_lists_map))

#Key Characteristics:

**Returns an iterator**:

map() returns a map object, which is an iterator. This means it doesn't immediately compute all results, but rather generates them on demand, making it memory-efficient for large iterables. You often need to convert it to a list or other data structure to see the full results.

**Functional programming concept**:

It aligns with functional programming principles by applying a function to a collection without modifying the original collection.



10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

Answer: The map(), filter(), and reduce() functions in Python are higher-order functions used for processing iterables, each serving a distinct purpose:

**map(function, iterable)**:

*Purpose*:

 Applies a given function to each item in an iterable and returns an iterator that yields the results.

*Output*:

 A new iterable (specifically, a map object) with the same number of elements as the input iterable, where each element is the result of applying the function.

*Example*:
 Transforming a list of numbers by squaring each one.

**filter(function, iterable)**:
*Purpose*: Constructs an iterator from elements of an iterable for which the function returns True.
*Output*: A new iterable (specifically, a filter object) containing only the elements from the input iterable that satisfy the condition defined by the function.
*Example*: Selecting only the even numbers from a list.

**reduce(function, iterable[, initializer])**:

*Purpose*: Applies a function of two arguments cumulatively to the items of an iterable from left to right, reducing the iterable to a single cumulative value.
*Note*: reduce() is part of the functools module and needs to be imported: from functools import reduce.
*Output*: A single value that is the result of the cumulative application of the function.
*Example*: Calculating the sum of all elements in a list.

*In summary*:
map() transforms elements, filter() selects elements, and reduce() aggregates elements into a single value.

```python
#5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

class FibonacciIterator:
    def __init__(self, max_terms):
        self.max_terms = max_terms
        self.count = 0
        self.a = 0
        self.b = 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.max_terms:
            raise StopIteration
        if self.count == 0:
            self.count += 1
            return 0
        elif self.count == 1:
            self.count += 1
            return 1
        else:
            self.count += 1
            self.a, self.b = self.b, self.a + self.b
            return self.a
fib = FibonacciIterator(10)

for num in fib:
    print(num)
```

```python
#6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

def powers_of_two(max_exponent):
    for exp in range(max_exponent + 1):
        yield 2 ** exp
for power in powers_of_two(5):
    print(power)
```

```python
#10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

def remove_vowels(input_str):
    vowels = 'aeiouAEIOU'
    return ''.join(filter(lambda char: char not in vowels, input_str))
text = "Hello, how are you?"
result = remove_vowels(text)
print("String without vowels:", result)

```
